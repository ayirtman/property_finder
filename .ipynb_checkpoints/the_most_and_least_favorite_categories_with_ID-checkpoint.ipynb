```python
import pandas as pd

#Importing dataset
popular_article = pd.read_csv("./articles_metadata.csv")

#Select category from dataset 
categoryID = popular_article['category_id']

array = []
batch_size = 10000

iteration_size = len(categoryID)//batch_size

#iterating categories with batch size. 
#Throughout iterating function prints the most and the least favorite 
# categories with ID and count number 
for i in range(iteration_size):
    array = []
    for j in range(batch_size):
        array.append(categoryID[i*batch_size+j])
        arrayCountValue = pd.Series(array)
    
    most_favorite_category_ID = arrayCountValue.value_counts().head(1)
    least_favorite_category_ID = arrayCountValue.value_counts().tail(1)
    
    print([most_favorite_category_ID, least_favorite_category_ID])
```
