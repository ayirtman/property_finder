```python
import numpy as np
from csv import writer 

# Path to click_sample dataset 
m_path = r"./clicks_sample.csv"

# Asked columns of the data set
clickEnvironment_df = pd.read_csv(m_path, usecols=['click_environment']) 
deviceGroup_df = pd.read_csv(m_path, usecols=['click_deviceGroup'])
operatingSystem_df = pd.read_csv(m_path, usecols=['click_os'])

# Empty array to append later
m_arr =  np.array([])
m_arr2 =  np.array([])
m_arr3 =  np.array([])

# Convert dataframes to numpy for calculating
temp = clickEnvironment_df.to_numpy()
temp2 = deviceGroup_df.to_numpy()
temp3 = operatingSystem_df.to_numpy()

# imaginary time interval for taking most average values of columns
clickDf_step = 30

def distribution_of_click(file_name, list_of_elem, list_of_elem_1,list_of_elem_2):
    evironment = ""
    device = ""
    os = ""

# Etiquete for the values
# They will be added on csv file
    if list_of_elem > 3:
        environment = "Web"
    elif list_of_elem > 2: 
        environment = "AMP"
    elif list_of_elem > 1:
        environment = "Mobile App"
    elif list_of_elem > 0:
        environment = "FaceBook Instant Article"
    else:
        environment = "Null Data"
        
    if list_of_elem_1 > 4:
        device = "Desktop"
    elif list_of_elem_1 > 3:
        device = "Mobile"
    elif list_of_elem_1 > 2:
        device = "Empty"
    elif list_of_elem_1 > 1:
        device = "TV"
    elif list_of_elem_1 > 0:
        device = "Tablet"
    else:
        device = "Null Data"
        
        
    if list_of_elem_2 > 22:
        os = "Solaris"
    elif list_of_elem_2 > 21:
        os = "Gentoo"
    elif list_of_elem_2 > 20:
        os = "webOS"
    elif list_of_elem_2 > 19:
        os = "Chromecast"
    elif list_of_elem_2 > 18:
        os = "Brew MP"
    elif list_of_elem_2 > 17:
        os = "Android"
    elif list_of_elem_2 > 16:
        os = "Firefox OS"
    elif list_of_elem_2 > 15:
        os = "BlackBerry OS"
    elif list_of_elem_2 > 14:
        os = "Symbian OS"
    elif list_of_elem_2 > 13:
        os = "Debian"
    elif list_of_elem_2 > 12:
        os = "Chrome OS"
    elif list_of_elem_2 > 11:
        os = "tvOS"
    elif list_of_elem_2 > 10:
        os = "ATV OS X"
    elif list_of_elem_2 > 9:
        os = "FireHbbTV"
    elif list_of_elem_2 > 8:
        os = "Samsung"
    elif list_of_elem_2 > 7:
        os = "Mac OS X"
    elif list_of_elem_2 > 6:
        os = "Windows"
    elif list_of_elem_2 > 5:
        os = "Windows Mobile"
    elif list_of_elem_2 > 4: 
        os = "Windows Phone"
    elif list_of_elem_2 > 3:
        os = "Android"
    elif list_of_elem_2 > 2:
        os = "iOS"
    elif list_of_elem_2 > 1:
        os = "Other"
    else:
        os = "Null Data"
        
        
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow([list_of_elem, environment, list_of_elem_1, device, list_of_elem_2, os])
        


    
    
# Take average values for adding to the csv file        
for i in range(1, len(temp), clickDf_step):
    arr = np.append(m_arr,np.array([np.average(temp[i:i + clickDf_step])]))
    arr2 = np.append(m_arr2,np.array([np.average(temp2[i:i + clickDf_step])]))
    arr3 = np.append(m_arr3,np.array([np.average(temp3[i:i + clickDf_step])]))
    
    distribution_of_click('event.csv',arr,arr2,arr3)
        
```
